# Questão 1 - Vale até 1 ponto
Em uma região desértica, pesquisadores estão estudando o crescimento de uma espécie rara de cacto. O crescimento do cacto ao longo do tempo é modelado pela função $y = x^5(2 - x)$, onde $y$ representa a altura do cacto em metros e $x$ representa o tempo em anos desde o início do estudo. Os pesquisadores querem determinar a quantidade total de biomassa acumulada pelo cacto ao longo dos primeiros 2 anos. Sabemos, de estudos anteriores que a biomassa acumulada é diretamente proporcional (1:1) a área sob a curva de crescimento do cacto ao longo do tempo. Determine a quantidade total de biomassa acumulada nestes dois anos.



$Area = \int_{0}^{2} x^5(2-x)$
\
$Area = \int_{0}^{2} 2x^5 - x^6$
\
$Area = 2\int_{0}^{2} x^5 - \int_{0}^{2} x^6$
\
$Area = 2\left[\frac{1}{6}x^6\right]_{0}^{2} - \left[\frac{1}{7}x^7\right]_{0}^{2}$
\
$Area = \frac{128}{6} - \frac{128}{7}$
\
$Area= \frac{896}{42} - \frac{768}{42}$
\
$Area= \frac{128}{42}$

Dando aquela simplificada final.

$A = 3.\overline{047619}$

E agora bora demonstrar com o soberbo Python:

```python
import sympy as sp

x = sp.symbols('x')
f = x**5 * (2 - x)

x_min, x_max = 0, 2
area = sp.integrate(f, (x, x_min, x_max))
print(f'A área se aproxima de: {abs(area).evalf()}')
```

# Questão 3 - Vale até 1 ponto

A aceleração de um objeto é dada por $a(t) = \cos(\pi t)$, e sua velocidade no tempo $t = 0$ é $\frac{1}{2\pi}$. Usando apenas álgebra, encontre a distância total percorrida por este objeto nos primeiros $1,5$ segundos deste movimento. Dica: antes de calcular qualquer coisa verifique cuidadosamente a equação da aceleração. Não é necessário, mas talvez seja interessante, traçar um gráfico da aceleração e outro da velocidade para entender o problema. É necessário entender o movimento do objeto para saber quanto espaço ele realmente percorreu.

A aceleração $a(t)$ é dada por:

$$a(t) = \cos(\pi t)$$

Sabendo que a velocidade inicial $v(0) = \frac{1}{2\pi}$, podemos encontrar a velocidade $v(t)$ integrando a aceleração:

$$v(t) = \int a(t) dt = \int \cos(\pi t) dt$$

Substituindo $u = \pi t$ e $du = \pi dt$, temos:

$$v(t) = \int \cos(u) \frac{du}{\pi}$$

Integrando, obtemos:

$$v(t) = \frac{\sin(\pi t)}{\pi} + C$$

Sabendo que $v(0) = \frac{1}{2\pi}$, podemos encontrar a constante $C$:

$$v(0) = \frac{\sin(\pi 0)}{\pi} + C = 0 + C = \frac{1}{2\pi}$$

Portanto, $C = \frac{1}{2\pi}$ e:

$$v(t) = \frac{\sin(\pi t)}{\pi} + \frac{1}{2\pi}$$

Agora, podemos calcular a distância percorrida $d$ entre $t=0$ e $t=1.5$ segundos:

$$d = \int_{0}^{1.5} v(t) dt = \int_{0}^{1.5} \left(\frac{\sin(\pi t)}{\pi} + \frac{1}{2\pi}\right) dt$$

Separando a integral em duas partes, obtemos:

$$d = \frac{1}{\pi} \int_{0}^{1.5} \sin(\pi t) dt + \int_{0}^{1.5} \frac{1}{2\pi} dt$$

Integrando, obtemos:

$$d = \left[\frac{-\cos(\pi t)}{\pi^2} + \frac{t}{2\pi}\right]_{0}^{1.5}$$

Avaliando a expressão entre $t=0$ e $t=1.5$, obtemos:

$$d = \frac{1.5}{2\pi} + \frac{1}{\pi^2} - \frac{1}{\pi^2}$$

Considerando $\pi \approx 3.14$, obtemos:

$$d \approx 0.340$$

Portanto, em 1.5 segundos, a distância percorrida foi de aproximadamente 0.340 metros.
Sei que náo era necessario mas sobrou tempo entao contemple o grafico da aceleracao e da velocidade respectivamente:


```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 1.5, 400)

a = np.cos(np.pi * t)

plt.plot(t, a, label='$a(t) = cos(πt)$')
plt.xlabel('Tempo (s)')
plt.ylabel('Aceleração (m/s²)')
plt.title('Gráfico da Aceleração')
plt.grid(True)
plt.legend()
plt.show()
```
